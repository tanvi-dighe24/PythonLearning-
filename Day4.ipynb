```python
#1️⃣ for Loop – Iterate Over Sequences
#Python's for loop is used to repeat a block of code for each item in a sequence (list, string, tuple, etc.).
#Basic Syntax:
for variable in sequence:
 #block of code
```

```python
fruits = ['apple', 'banana', 'cherry']

for fruit in fruits:
    print(fruit)
```

```python
#✅ What is Indentation?
#Indentation means adding spaces or a tab at the beginning of a line to show that it belongs to a block of code — like inside a loop, function, or condition.

#In other programming languages (like C, Java), we use {} to define a block.
#But in Python, indentation replaces {}.
#🧠Rules to Remember:
#Indentation is usually 4 spaces or 1 tab.
#You must indent consistently in the block.
#Python will give an error if indentation is missing or mixed.
#🔎 Where Indentation is Needed?
#After for, while
#Inside if, else, elif
#Inside functions and classes
```

```python
#2️⃣ while Loop – Repeat Until False
#The while loop runs as long as a condition is True.
#🔹Basic Syntax:
#while condition:
    # code block
#✅ Example:
i = 1
while i <= 5:
    print(i)
    i += 1
```

```python
#3️⃣ break, continue, and pass
#break: Stops the loop entirely
#continue: Skips the current iteration and continues
#pass: Placeholder (does nothing)
#Example with break:
for i in range(10):
    if i == 5:
        break
    print(i) #01234

#Example with continue:
for i in range(5):
    if i == 2:
        continue
    print(i)  #01234

#Example with pass:
for i in range(3):
    pass  # Used when loop is empty for now #no output
```

```python
for i in range(10):
    if i == 5:
        break
    print(i)
```

```python
for i in range(5):
    if i == 2:
        continue
    print(i)
```

```python
#4️⃣ else With Loops
#Python allows an else block with loops. It executes only if the loop wasn't terminated by a break.
for i in range(5):
    print(i)
else:
    print("Loop completed")


```

```python
#But if there's a break:
for i in range(5):
    if i == 3:
        break
    print(i)
else:
    print("Won't print this")
```
